```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64;     x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate",     "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
```

```python
url="https://www.delish.com/cooking/recipe-ideas/a30149814/grapefruit-sidecar-recipe/"
```

```python
resp=requests.get(url, headers=headers) # we will make request from website to get response
```

```python
resp
```

```python
a=bs(resp.content, "lxml") # now we need a parser(lxml) that parse whole website content
```

```python
print(a.title) #it will show title to the page
```

```python
print(a.title.text) #only text no title tag
```

```python
print(a.p) #it will print first paragraph
```

```python
print(a.p.text.strip()) #print first paragraph text removing all the whitespace
```

```python
#print(a.find_all("p"))
```

```python
for paragraph in a.find_all("p"):
    print(paragraph.text.strip())
```

```python
for paragraph in a.find_all("p"):
    print(paragraph.string) # string dont allow child tags
```

```python
print(a.get_text()) # some website dont use paragraph<p></p> tags so get_text is better to use
```

```python
for urls in a.find_all("a"):   #parsing over all the tage having a
    print(urls)
```

```python
for urls in a.find_all("a"):   #parsing over all the tage having a
    print(urls.get("href"))   #we can get all the urls only having href
```
