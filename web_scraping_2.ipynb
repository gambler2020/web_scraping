```python
from bs4 import BeautifulSoup as bs
```

```python
import requests
```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64;     x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate",     "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
```

```python
urls="https://sportytell.com/net-worth/lionel-messi-net-worth/"
```

```python
resp=requests.get(urls, headers=headers)
```

```python
resp
```

```python
s=bs(resp.content, "lxml")
```

```python
for n in s.find_all("a"): # all the links
    print(n.get("href"))
```

```python
s.body # we are accessing only body part
```

```python
body=s.body
```

```python
body
```

```python
for paragraph in body.find_all("p"): #inside body i want to get all the paragraph with tag p
    print(paragraph.text.strip().replace("Table of Contents", ""))
```

```python
for div in s.find_all("div"):
    print(div.text)
```
